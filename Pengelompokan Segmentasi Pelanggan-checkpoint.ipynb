### LOADING DATA

```python
#Loading the dataset
data = pd.read_csv("marketing_campaign.csv", sep="\t")
print("Jumlah titik data:", len(data))
data.head()
```

### DATA CLEANING

```python
data.info()
```

**Dari output di atas, dapat disimpulkan bahwa:**

* Ada nilai yang hilang dalam pendapatan
* Dt_Customer yang menunjukkan tanggal pelanggan bergabung dengan database tidak diurai sebagai DateTime

Pertama-tama, untuk nilai yang hilang, saya hanya akan menghilangkan baris yang memiliki nilai pendapatan yang hilang.

```python
#To remove the NA values
data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ['Marital_Status', 'Education', 'Dt_Customer']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

Membuat fitur **("Customer_For")** dari jumlah hari pelanggan mulai berbelanja di toko relatif terhadap tanggal terakhir yang tercatat

# EVALUASI MODEL

```python
#Plotting countplot cluster
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data["Clusters"], palette= pal)
pl.set_title("Distribusi Cluster")
plt.show()
```

```python
pl = sns.scatterplot(data = data,x=data["Spent"], y=data["Income"],hue=data["Clusters"], palette= pal)
pl.set_title("Profil Klaster Berdasarkan Pendapatan dan Belanja")
plt.legend()
plt.show()
```

```python
plt.figure()
pl=sns.swarmplot(x=data["Clusters"], y=data["Spent"], color= "#CBEDDD", alpha=0.5 )
pl=sns.boxenplot(x=data["Clusters"], y=data["Spent"], palette=pal)
plt.show()
```

```python
data["Total_Promosi"] = data["AcceptedCmp1"]+ data["AcceptedCmp2"]+ data["AcceptedCmp3"]+ data["AcceptedCmp4"]+ data["AcceptedCmp5"]
#Plotting jumlah total kampanye yang diterima.
plt.figure()
pl = sns.countplot(x=data["Total_Promosi"],hue=data["Clusters"], palette= pal)
pl.set_title("Jumlah Promosi Diterima")
pl.set_xlabel("Jumlah Total Promosi yang Diterima")
plt.show()
```

```python
plt.figure()
pl=sns.boxenplot(y=data["NumDealsPurchases"],x=data["Clusters"], palette= pal)
pl.set_title("Jumlah Penawaran yang Dibeli")
plt.show()
```

```python
Places =["NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases",  "NumWebVisitsMonth"] 

for i in Places:
    plt.figure()
    sns.jointplot(x=data[i],y = data["Spent"],hue=data["Clusters"], palette= pal)
    plt.show()
```
